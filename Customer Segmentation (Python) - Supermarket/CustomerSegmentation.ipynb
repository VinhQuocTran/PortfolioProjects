# 4. Customer Segmentaion

**Import needed libraries for cluster analysis model**

We will use K-Mean algorithm to group similar customers together and discover underlying pattern

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.colors
import random
```

## a. Feature Standardization

K-Mean often uses Euclidean Distance to form a group, if we do not standarize before feeding numerical features to ML algorithm then our model will be dominated or heavily biased by a feature with large scale <br>
For example: Income column has larger value than the rest, so it will be dominated in model if we don't perform standarization  

```python
sc = StandardScaler()
df_edited=df_edited.select_dtypes(include='number')
df_edited_scaled = pd.DataFrame(sc.fit_transform(df_edited),columns = df_edited.columns,index=df_edited.index)
df_edited_scaled
```

We can see that the data has been scaled successfully

## b. Feature Selection

Feature Selection is one of the most important tasks in customer segmentation problem. Different features affect clusters differently, some features are vital for clusters while the others may not be so important as you think <br> <br>
In this section, I will choose features based on several common characteristics of customer like:
- Demographic: includes personal information such as income, gender, race, nationality, occupation,...
- Geographic: includes climate, city or country, population density,..
- Psychographic: includes customer’s personality and preferences like attitudes, opinions, values, interests, lifestyle and motivations
- Behavioral: The way a customer behaves affects how they interact with businesses, includes purchasing habits, brand loyalty,..

```python
df_feats=df_edited_scaled[['Income', 'Days_is_customer', 'avg_check']].copy()
```

Based on 4 characteristics above, I decided to choose Income (Demographic), Days_is_customer (Behavioral), avg_check (Psychographic) columns as features for K-Mean algorithm

## c. Find optimal number of clusters by Silhoutte Method

The number of clusters plays an important role in customer segmentation, we don't want to have too much (hard to focus) or too little (lack of details) groups of customer <br>
In order to perform well in clustering, we want to minimize the distance between datapoints within a cluster to its centroid (increase compactness of cluster) and there are 2 main techniques that we can help us with that: The elbow method and Silhoutte method. If you want to find out more about these 2 methods, just click the link below <br>
https://vitalflux.com/elbow-method-silhouette-score-which-better/#:~:text=The%20major%20difference%20between%20elbow,high%2Dlow%20differences%2C%20etc.

```python
# Calculate Silhoutte score for each number of clusters
scoreList=[]
n=10
clusterNum=list(range(2,n))
for n_clusters in range(2,n):
    km = KMeans(n_clusters=n_clusters,init = 'k-means++')
    preds = km.fit_predict(df_feats)
    labels=km.labels_
    #centers = clusterer.cluster_centers_

    score = silhouette_score(df_feats, preds)
    scoreList.append(score)
    print("For n_clusters = {}, silhouette score is {})".format(n_clusters, round(score,3)))
```

```python
plt.plot(clusterNum,scoreList)
plt.xlabel('Number of cluster')
plt.ylabel('Silhoustte score')
```

**For the above-given graph, the optimal value for cluster would be any number between 2 to 5**. As we can see after 5, the more cluster you add the less Silhoutte score you would get. Furthermore, 2 to 5 clusters would be the ideal number for groups of customer

## d. Model building and visualization

### Model building and clusters's distribution

```python
# Number of clusters
k=4
km = KMeans(n_clusters=k,init='k-means++',random_state=42)

# Fit and predict label
preds = km.fit_predict(df_feats)
df_edited['Customer_Type']=preds
df_edited['Customer_Type'].value_counts().sort_index()
```

```python
def addlabels(x,y):
    total=sum(y)
    for i in range(len(x)):
        myStr=str(y[i])+' | '+str(round(y[i]*100/total,1))+'%'
        plt.text(i,y[i]//2,myStr,ha = 'center')
```

```python
clusters=list(range(k))
y=df_edited['Customer_Type'].value_counts().sort_index()
colors=('red','yellow','green','blue','gray')
plt.figure(figsize = (9,6))
plt.bar(clusters,y,color=colors,alpha=0.8)
plt.xticks(clusters)
plt.xlabel('Cluster no. ')
plt.ylabel('Count')
plt.title('Customer Segmentation')
addlabels(clusters, y)
```

### K-Mean Cluster 3D Plot 

```python
# You can create n random color use this function
def createRandomColor(n):
    number_of_colors  = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(n)]
    return number_of_colors

# Or if you want don't want random, just use this color palette
colors=['#FAD3AE', '#855E46', '#FE800F', '#890000']
cmap = matplotlib.colors.ListedColormap(colors)
```

```python
#Plotting the clusters
x=df_edited['Income']
y=df_edited['Days_is_customer']
z=df_edited['avg_check']

fig = plt.figure(figsize=(15,12))
ax = plt.subplot(111, projection='3d', label="bla")
sc=ax.scatter(x, y, z, s=40, c=df_edited['Customer_Type'], marker='o', cmap = cmap )
ax.set_title("The Plot Of The Clusters")
ax.set_xlabel('Income',fontsize=15)
ax.set_ylabel('Days is customer',fontsize=15)
ax.set_zlabel('Average Per Check',fontsize=15)
plt.legend(*sc.legend_elements())
```

### Naming Customer Segments

After finishing clustering, we should give each cluster a meaningful name to identify between them, this could be based on their characteristic features like Income, spending habit,...

```python
features=['Income', 'Days_is_customer', 'avg_check']
for feat in features:
    y=df_edited.groupby('Customer_Type').mean()[feat]
    plt.bar(clusters,y)
    plt.xticks(clusters)
    plt.xlabel('Cluster no.')
    plt.ylabel(feat)
    plt.show()
```

```python
df_edited["Age_Group"] = pd.cut(df_edited["Age"],  bins=[1,30,45,60,75],include_lowest=True)
prf = ["Children", "Age_Group"]
for i in prf:
    plt.figure(figsize=(13,8))
    sns.swarmplot(x=i, y="Expenses", hue="Customer_Type", palette=colors ,data=df_edited,size=2.5)
```

- Cluster **number 2**: These are lifelong customers, they have the highest average money per check and income, so they are **Elite Clients**
- Cluster **number 0**: Pretty much the same as 2th cluster, but the only thing is that they don't spend as much as them, Their average money per check, income are ranked second, so they are **Good Clients**
- Cluster **number 1**: would be **Potential Good Clients**
- Cluster **number 4**: would be **Ordinary Clients** because their features are the lowest among 4 clusters
